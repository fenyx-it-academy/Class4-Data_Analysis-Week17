Bir dongu icerisinde random olacak sekilde iki tane 20x20’luk matris uretin ve bu matrislerin farklarini alin. 
Ve fark matrisinin diagonali, -0.1 ile 0.1 arasinda olana kadar bu islemi tekrarlayin. 
Istenilen matris bulundugunda program dursun ve toplam kac dongunun kuruldugunu, ne kadar sure icinde buldugunu ve 
istenilen matrisi print ile birlikte ekrana yazdiriniz... 

Tavsiye: 20 x 20 luk matrisin bulunmasi saatler surebilir. 
Bu yuzden algoritmanizin dogrulugunu test etmek icin once 4x4, 5x5 gibi kucuk matrislerde deneyebilirsiniz. 
Ve matris sayisisini arttirarak en son 20x20 u deneyebilirsiniz.

```python
import time
loop = 0
t1 = time.time()

while True:
    loop+=1
    matrix = [np.random.rand(5,5) for _ in range(2)]
    
    m1 = matrix[0]
    m2 = matrix[1]
    
    diag = abs(np.diag(m1-m2))
    if len(list(filter(lambda x: x>0.1, diag)))==0:
        break
        
t2 = time.time()
time_diff = t2-t1
```
